```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
import re
```

```python
text = re.sub('[^a-zA-Z\d]'," ", text)
```

```python
text
```

```python
stopwords = list(STOP_WORDS)
```

```python
nlp=spacy.load("en_core_web_sm")
```

```python
doc=nlp(text)
```

```python
tokens = [token.text for token in doc]
print(tokens[:20])
len(tokens)
```

```python
punctuation
```

```python
punctuation = punctuation +"\n"+'  '+' '
```

```python
punctuation
```

```python
word_frequencies = {}
for word in doc:
    if word.text.lower() not in stopwords:
        if word.text.lower() not in punctuation:
            if word.text not in word_frequencies.keys():
                word_frequencies[word.text]=1
            else:
                word_frequencies[word.text]+=1
```

```python
print(word_frequencies)
```

```python
max_frequency = max(word_frequencies.values())
```

```python
max_frequency
```

```python
#normalize the frequency
```

```python
import matplotlib.pyplot as plt
```

```python
k = plt.hist(list(word_frequencies.values()))
```

```python
#normalization
for word in word_frequencies.keys():
    word_frequencies[word]=word_frequencies[word]/max_frequency
```

```python
print(word_frequencies)
```

```python
#normalised
k = plt.hist(list(word_frequencies.values()))
```

```python
import pandas as pd
```

```python
a=pd.Series(list(word_frequencies.keys()))
b=pd.Series(list(word_frequencies.values()))
```

```python
topwords=pd.concat([a,b],axis=1)
topwords.columns=["words","frequency"]
topwords.sort_values(by="frequency", ascending=False).head(20)
```

```python
list(word_frequencies.keys())
```

```python
#sentence tokenization
```

```python
sentence_tokens = [sent for sent in doc.sents]
print(sentence_tokens)
```

```python
#aplly back word frequencies to sentences
sentence_scores ={}
for sent in sentence_tokens:
    for word in sent:
        if word.text.lower() in word_frequencies.keys():
            if sent not in sentence_scores.keys():
                sentence_scores[sent]=word_frequencies[word.text.lower()]
            else:
                sentence_scores[sent]+=word_frequencies[word.text.lower()]
        
           
```

```python
print(sentence_scores)
```

```python
#make summary with maxium scores
```

```python
from heapq import nlargest
```

```python
select_length = int(len(sentence_tokens) *0.3)
select_length
```

```python
#syntax = nlargest(n, iterable, key=None)
summary = nlargest(select_length,sentence_scores, key= sentence_scores.get)
```

```python
summary
```

```python
final_summary=[word.text for word in summary]
```
